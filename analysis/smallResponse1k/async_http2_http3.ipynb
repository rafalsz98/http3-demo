```python
import pandas as pd
import numpy as np
async_http2 = pd.read_csv('http2asynctest.csv', delimiter=";")
async_http3 = pd.read_csv('http3asynctest.csv', delimiter=";")

mean_http2 = async_http2.mean()
min_http2 = async_http2.min()
max_http2 = async_http2.max()

mean_http3 = async_http3[async_http3["http_version"] == 3].mean()
min_http3 = async_http3[async_http3["http_version"] == 3].min()
max_http3 = async_http3[async_http3["http_version"] == 3].max()

print("All HTTP/2 requests: " + str(async_http2.shape[0]))
print("Mismatched HTTP/2 requests: " + str(async_http2[async_http2["http_version"] != 2].shape[0]))
print("Failed HTTP/2 requests: " + str(async_http2[async_http2["response_code"] != 200].shape[0]))
print("")
print("All HTTP/3 requests: " + str(async_http3.shape[0]))
print("Mismatched HTTP/3 requests: " + str(async_http3[async_http3["http_version"] != 3].shape[0]))
print("Failed HTTP/3 requests: " + str(async_http3[async_http3["response_code"] != 200].shape[0]))
pd.DataFrame(
  {
    "HTTP/2 Min": min_http2.values, 
    "HTTP/2 Mean": mean_http2.values, 
    "HTTP/2 Max": max_http2.values, 
    "HTTP/3 Min": min_http3.values,
    "HTTP/3 Mean": mean_http3.values,
    "HTTP/3 Max": max_http3.values
  },
  mean_http2.keys(),
  [
    "HTTP/2 Min",
    "HTTP/2 Mean",
    "HTTP/2 Max",
    "HTTP/3 Min",
    "HTTP/3 Mean",
    "HTTP/3 Max"
  ]
)
```
